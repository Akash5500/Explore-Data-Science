### Data visualization tutorial

Data Visualization is a technique of presenting data graphically or in a pictorial format which helps to understand large quantities of data very easily. This allows decision-makers to make better decisions and also allows identifying new trends, patterns in a more efficient way. 

### major libraries which help us in visualization are follows

* __Matplotlib__
* __Seaborn__
* __Plotly__

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig = plt.figure()

# Create a circular subplot
circle_axes = fig.add_axes([0.1, 0.3, 0.4, 0.4], aspect='equal', frameon=False)

# Add an Ellipse patch to simulate a circle
circle = patches.Ellipse((0.5, 0.5), 1, 1, edgecolor='black', facecolor='lightgoldenrodyellow')
circle_axes.add_patch(circle)

# Customize tick labels and tick lines
for label in circle_axes.xaxis.get_ticklabels():
    label.set_color('green')
    label.set_rotation(25)
    label.set_fontsize(16)

for line in circle_axes.yaxis.get_ticklines():
    line.set_color('yellow')
    line.set_markersize(5)
    line.set_markeredgewidth(3)

# Show the plot
plt.show()
```

```python
# Implementation of matplotlib function 
import matplotlib.pyplot as plt 
import matplotlib.tri as mtri 
import numpy as np 
	
# Create triangulation. 
x = np.asarray([0, 1, 2, 3, 0.5, 
				1.5, 2.5, 1, 2, 
				1.5]) 

y = np.asarray([0, 0, 0, 0, 1.0, 
				1.0, 1.0, 2, 2, 
				3.0]) 

triangles = [[0, 1, 4], [1, 5, 4], 
			[2, 6, 5], [4, 5, 7], 
			[5, 6, 8], [5, 8, 7], 
			[7, 8, 9], [1, 2, 5], 
			[2, 3, 6]] 

triang = mtri.Triangulation(x, y, triangles) 
z = np.cos(3 * x) * np.cos(6 * y)+np.sin(6 * x) 
	
fig, axs = plt.subplots() 
t = axs.tricontourf(triang, z) 
axs.tricontour(triang, z, colors ='white') 
fig.colorbar(t) 

fig.suptitle('matplotlib.figure.Figure.colorbar() ', fontweight ="bold") 
print(fig.get_figwidth())
print(fig.get_figheight())
plt.show() 

```

###  different ways to creat a subplots
* __Gridspec__
* __gridspec_kw__

```python
# Gridspec

# gridspec.GridSpec(ncols=2, nrows=2, width_ratios=[2, 1],wspace=0.5,hspace=0.5,height_ratios=[1,2])

import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
 
# create a figure
fig= plt.figure()   # fig= plt.figure(figsize=(8,8)) 

fig.set_figheight(8)
fig.set_figwidth(8)

spec=gridspec.GridSpec(nrows=2,ncols=2,  wspace=0.5,hspace=0.5,width_ratios=[2,1],height_ratios=[2, 1])

x = np.arange(0, 10, 0.1)
y = np.cos(x)

ax0=fig.add_subplot(spec[0])
ax0.plot(x,y)
ax1 = fig.add_subplot(spec[1])
ax1.plot(x, y)
ax2 = fig.add_subplot(spec[2])
ax2.plot(x, y)
ax3 = fig.add_subplot(spec[3])
ax3.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# setting different parameters to adjust each grid
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(4,4),facecolor='y',
                       gridspec_kw={
                        'width_ratios': [3, 1],
                        'height_ratios': [3, 3],
                       'wspace': 0.4,
                       'hspace': 0.4})
 
 
# initializing x,y axis value
x = np.arange(0, 10, 0.1)
y = np.tan(x)

ax[0][0].plot(x, y)
 
# ax[0][0] will take 0th position in
# geometry(Grid we created for subplots)
ax[0][1].plot(x, y)
 
# ax[0][0] will take 0th position in
# geometry(Grid we created for subplots)
ax[1][0].plot(x, y)
 
# ax[0][0] will take 0th position in
# geometry(Grid we created for subplots)
ax[1][1].plot(x, y)
 
plt.show()
```
