### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(1000)

# print(data)

plt.hist(data, bins=30, color='skyblue', edgecolor='black',align='left')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

```python
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

np.random.seed(23685752)
N_points = 10000
n_bins = 20

x = np.random.randn(N_points)
y = .8 ** x + np.random.randn(10000) + 25
legend = ['distribution']
 
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7), 
                        tight_layout = True)

for s in ['top', 'bottom', 'left', 'right']: 
    axs.spines[s].set_visible(False) 

axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')

axs.grid(color ='grey', 
        linestyle ='-.', linewidth = 0.5, 
        alpha = 0.6) 

axs.xaxis.set_tick_params(pad = 5) 
axs.yaxis.set_tick_params(pad = 10) 

fig.text(0.9, 0.15, 'Akash Hudedh', 
         fontsize = 12, 
         color ='red',
         ha ='right',
         va ='bottom', 
         alpha = 0.3) 

axs.hist(x, bins = n_bins)
```

```python
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(12,4))
ax[0].hist(data1,bins=30,color='lightgreen',edgecolor='k',label='Data1')
ax[0].grid('-.',linewidth=0.5,color='grey',alpha=0.4)
ax[0].xaxis.set_ticks_position('none')
ax[0].yaxis.set_ticks_position('none')
ax[0].legend(frameon=False)

ax[1].hist(data2,bins=30,color='lightblue',edgecolor='k',label='Data2')
ax[1].grid('-.',linewidth=0.5,color='grey',alpha=0.4)
ax[1].xaxis.set_ticks_position('none')
ax[1].yaxis.set_ticks_position('none')
ax[1].legend(frameon=False)

for s in ['top', 'bottom', 'left', 'right']: 
    ax[0].spines[s].set_visible(False) 
    ax[1].spines[s].set_visible(False) 
    

ax[0].set_xlabel('Values')
ax[0].set_ylabel('Frequency')
ax[1].set_xlabel('Values')
ax[1].set_ylabel('Frequency')

fig.tight_layout()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
colors = ["orange", "cyan", "brown",
          "grey", "indigo", "beige"]
wp = {'linewidth': 1, 'edgecolor': "green"}


plt.pie(data,
        labels=cars,
        autopct='%1.2f%%',
        explode=explode,
        shadow=True, 
        colors=colors,
        startangle=0,
        wedgeprops=wp,
       textprops=dict(color='b'))
plt.legend(bbox_to_anchor=(1,1,0.5,0),frameon=True,labelcolor='b',labelspacing=2,shadow=True,ncols=3,columnspacing=0.5)
plt.show()
```

```python
years = [2016, 2017, 2018, 2019, 2020] 
profit = [15, 19, 35, 14, 17] 
wp={
    'edgecolor':'black',
    'antialiased': True
}
explode=[0.1,0.0,0.3,0.4,0.2]

plt.pie(profit,
       labels=years,
       startangle=90,
       autopct='%1.1f%%',
       wedgeprops=wp,
       pctdistance=0.5,
       labeldistance=1.1,
       radius=1,
       explode=explode,
       shadow=True,
       frame=False)
plt.axis('equal')
plt.show()
```
