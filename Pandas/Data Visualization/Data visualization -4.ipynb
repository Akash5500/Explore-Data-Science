### 3D PLOT

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)


ax.plot3D(x,y,z,'g')
ax.set_title('First 3D Plot')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

c=x+y

ax.scatter(x,y,z ,c=c)
```

```python
import random
x=np.random.randint(100,size=(100))
y=np.random.randint(80,size=(100))
z=np.random.randint(60,size=(100))

ax=plt.axes(projection='3d')
ax.scatter3D(x,y,z,c=x,cmap='bone')
plt.show()
```

```python
z = 4 * np.tan(np.random.randint(10, size =(500))) + np.random.randint(100, size =(500))
x = 4 * np.cos(z) + np.random.normal(size = 500)
y = 4 * np.sin(z) + 4 * np.random.normal(size = 500)
 
# Creating figure
fig = plt.figure(figsize = (16, 9))
ax=plt.axes(projection='3d')
              
scc=ax.scatter3D(x,y,z,
            alpha=0.7,
            c=x,
            cmap='hsv',
            marker='^')
plt.colorbar(scc,shrink=0.5,location='right')
```

###  Surface plot

```python
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='nipy_spectral')
```

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T # transpose
z = (np.sin(x **2) + np.cos(y **2) )
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
```

### Wireframes graph

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
 
# x and y axis
x = np.linspace(-1, 5, 10)
y = np.linspace(-1, 5, 10)
  
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection ='3d')

ax.plot_wireframe(X,Y,Z,color='g')
```

```python
# importing modules
from mpl_toolkits.mplot3d import axes3d
from matplotlib import pyplot 

# creating the visualization
fig = pyplot.figure()
wf = fig.add_subplot(111, projection='3d')
x, y, z = axes3d.get_test_data(0.05)
wf.plot_wireframe(x,y,z, 
                  rstride=2, 
                  cstride=2,
                  color='green')

# displaying the visualization
wf.set_title('Example 1')
pyplot.show()
```

### Contour Graphs

```python
def function(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
 
 
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
 
X, Y = np.meshgrid(x, y)
Z = function(X, Y)
 
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z)
```
