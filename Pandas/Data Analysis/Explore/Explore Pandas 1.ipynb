### syntax for replace values inside table(R*C)

DataFrame.replace(to_replace=None, value=None, inplace=False, limit=None, regex=False, method=’pad’, axis=None) 

```python
import pandas as pd
import numpy as np

data=pd.read_csv('nba.csv', index_col='Name')
df=pd.DataFrame(data)

# df['Name']=df['Name'].str.lower()
# df['Name'].apply(lambda x:x.lower())
df_lower=df.applymap(lambda x : x.casefold() if isinstance(x,str) else x)
df=df_lower
df.index=df.index.str.lower()
df.columns=df.columns.str.upper()
df.rename(columns={'TEAM':'TEAM1'},inplace=True)
df.rename(index={'avery bradley':'ak'},inplace=True)
df.columns=df.columns.str.replace('AGE','age1')
df.replace(to_replace=['boston celtics','brooklyn nets'],value='boston celtics1',inplace=True)
df.replace(to_replace=np.nan,value=-99999,inplace=True)
df.replace(to_replace='celtics1',value='123',inplace=True,regex=True)
df.head(n=60)
```
